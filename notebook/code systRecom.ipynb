### Feature Ingenering

```python
# 1. Sélection des colonnes nécessaires
columns_to_keep = ['title', 'genres', 'production_countries', 'keywords', 'release_date']
data = data[columns_to_keep]

data
```

```python
# prendre que 3 termes de genres
data['genres'] = data['genres'].str.split(',').apply(lambda x: ', '.join(x[:3]) if isinstance(x, list) else '')

# prendre que 2 termes des mots clés
data['keywords'] = data['keywords'].str.split(',').apply(lambda x: ', '.join(x[:2]) if isinstance(x, list) else '')

data
```

```python
# Extraire l'année directement en utilisant str et split
data['release_year'] = data['release_date'].str.split('-').str[0]

# Supprimer la colonne 'release_date'
data.drop(columns=['release_date'], inplace=True)

data
```

####  Creation d'une colonne qui combine toutes les autres colonnes 'cat_features'

```python
data['cat_features']=data[data.columns].apply(lambda x: ''.join(x), axis=1)

data.head()
```

### vectorisation 

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english', min_df=20)
word_matrix = vectorizer.fit_transform(data['cat_features'])
word_matrix.shape
```

### Calcul de cosinus de similaté

Le calcul de cosinus de similarité permet de....

```python
from sklearn.metrics.pairwise import cosine_similarity

sim = cosine_similarity(word_matrix)
```

Construction de la fonction pour le systeme de recommandation des films en question 

### Fonction du systeme recommandation

```python
def get_recom(titre, data, sim, count=10):
    """
    Obtenir des recommandations basées sur la similarité cosinus.
    
    Parameters:
    - titre : str, titre du film pour lequel des recommandations sont recherchées.
    - data : DataFrame, contenant au moins une colonne 'title'.
    - sim : matrice de similarité (numpy array ou DataFrame).
    - count : int, nombre de recommandations souhaitées (par défaut 10).
    
    Returns:
    - list : titres des films recommandés.
    """
    # Trouver l'indice correspondant au titre
    index = data[data['title'].str.lower() == titre.lower()].index
    
    # Vérifier si le titre existe
    if len(index) == 0:
        return []
    
    # Récupérer l'indice
    idx = index[0]
    
    # Vérifier si l'indice est dans les limites de la matrice
    if idx >= len(sim):
        return []
    
    # Obtenir les similarités de la ligne correspondante
    similarities = list(enumerate(sim[idx]))
    
    # Trier par score de similarité décroissant
    recommandations = sorted(similarities, key=lambda x: x[1], reverse=True)
    
    # Obtenir les meilleurs résultats, en excluant l'élément lui-même
    top_rec = recommandations[1:count + 1]
    
    # Générer la liste des titres
    titres = []
    for rec in top_rec:
        rec_idx = rec[0]  # Indice de la recommandation
        if rec_idx < len(data):  # Vérifier les limites
            titre_rec = data.iloc[rec_idx]['title']
            titres.append(titre_rec)
    
    return titres
```

```python
# Exemple d'application
recommandations = get_recom(
    titre='the avengers',
    data=data,  # DataFrame contenant les informations sur les films
    sim=sim,    # Matrice de similarité cosinus
    count=9     # Obtenir les 5 meilleures recommandations
)

recommandations
```
